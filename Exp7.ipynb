```python
#Write  python program for fibonacci series


n = int(input("Enter number of terms: "))
a = 0
b = 1
i = 1
while i <= n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
# Reverse a number


num = int(input("Enter a number: "))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
  print("Reverse number =",rev)
```

```python
# Check palindrome number/string


num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0 :
  rev = rev * 10 + num % 10
  num //= 10
if temp == rev :
  print("Palindrone")
else:
  print("Not Palindrone")
```

```python
# check palindrome number


st = input("Enter the string:")
rev = st[::-1]
if st==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
  if is_palindrome:
    print("yes")
  else:
    print("no")
```

```python
# Count digits in number


num = int(input("Enter a number: "))
count = 0

while num > 0:
  count += 1
  num = num// 10

print("Number of digits =",count)
```

```python
# search an element in  a list


nums = [10, 20, 30, 40, 50]
key =  int(input("Enter elemnt to search: "))
i = 0
while i < len(nums):
  if nums[i] == key:
    print("Element found at index",i)
    break
  i += 1
else:
  print("Element not found")
```
